```python
# SPLIT the entire set into four different ones, based on the value in the group column
g1 = fullset.loc[fullset['group']=='I']
g2 = fullset.loc[fullset['group']=='II']
g3 = fullset.loc[fullset['group']=='III']
g4 = fullset.loc[fullset['group']=='IV']
```

```python
from sklearn.linear_model import LinearRegression # pip install scikit-learn
import matplotlib.pyplot as plt #pip install matplotlib
```

```python
model = LinearRegression()
```

```python
g1x = g1.x.values.reshape(11, 1)
g1y = g1.y.values.reshape(11, 1)
model.fit(g1x, g1y)
model.intercept_
```

```python
plt.scatter(g1x, g1y, color = 'green', )
plt.plot(g1x, model.predict(g1x), color = 'red',linewidth=1)
plt.title('One')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
g2x = g2.x.values.reshape(11, 1)
g2y = g2.y.values.reshape(11, 1)
model.fit(g2x, g2y)
model.intercept_
plt.scatter(g2x, g2y, color = 'green')
plt.plot(g2x, model.predict(g2x), color = 'red',linewidth=1)
plt.title('Two')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
